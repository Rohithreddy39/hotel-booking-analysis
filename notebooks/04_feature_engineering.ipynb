# Feature Engineering â€” Hotel Booking Analysis

This notebook creates new analytical features for EDA and dashboards.


```python
#Load the final cleaned dataset
import pandas as pd 

df=pd.read_csv("../data/python_cleaned/hotel_bookings_cleaned_final.csv")
df.shape
```

```python
#Create key features

#Total Guests
df['total_guests']=df['adults']+df['children']+df['babies']

#Total nights (alraedy present, but verify)
df['total_nights'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']

#Arrival date (convert date to datetime)
df['arrival_date'] = pd.to_datetime(
    df['arrival_date_year'].astype(str) + "-" +
    df['arrival_date_month'].astype(str) + "-" +
    df['arrival_date_day_of_month'].astype(str),
    errors='coerce'
)


#Booking Month/Year
df['arrival_month']=df['arrival_date'].dt.month
df['arrival_year']=df['arrival_date'].dt.year

#Family Flag
df['is_family']=(df['children']+df['babies']>0).astype(int)


#Revenue estimate (simple): adr * nights
df['revenue']=df['adr'] * df['total_nights']
df.loc[df['is_canceled'] == 1,"revenue"] = 0
```

```python
#Validate new features
df[['total_guests','total_nights','arrival_date','is_family','revenue']].describe(include='all')
```

```python
#Check for missing values
df['arrival_date'].isna().sum()
```

```python
#Save the featured dataset
df.to_csv("../data/python_cleaned/hotel_bookings_features.csv", index=False)
```
