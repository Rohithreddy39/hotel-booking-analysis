# Bivariate Exploratory Data Analysis (EDA)

This notebook explores relationships between key variables in the hotel booking dataset.


```python
#Load the Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("../data/python_cleaned/hotel_bookings_features.csv")
df.shape
```

Core Business Questions + Plots

```python
#1️⃣ Are cancellations higher in City Hotel or Resort Hotel?
plt.figure(figsize=(8,5))

sns.barplot(x="hotel",y="is_canceled",data=df,estimator=lambda x: sum(x)/len(x))

plt.title("Cancellation Rate by hotel type.")
plt.ylabel("Cancellation Rate")
plt.show()
```

```python
#2️⃣ Does lead time affect cancellation?
plt.figure(figsize=(8,5))

sns.boxplot(x="is_canceled",y="lead_time",data=df)
plt.title("Lead Time vs Cancellation")
plt.xlabel("Is canceled (0=No, 1=Yes)")
plt.show()
```

```python
#3️⃣ Does higher ADR lead to more cancellations?
plt.figure(figsize=(8,5))

sns.boxplot(x="is_canceled",y="adr",data=df)
plt.title("ADR vs Cancellation.")
plt.show()
```

```python
#4️⃣ Do families cancel more than non-families?
plt.figure(figsize=(6,4))
sns.barplot(x="is_family", y="is_canceled", data=df, estimator=lambda x: sum(x)/len(x))
plt.title("Cancellation Rate: Family vs Non-Family")
plt.xlabel("Is Family (0 = No, 1 = Yes)")
plt.ylabel("Cancellation Rate")
plt.show()
```

```python
#5️⃣ Does market segment affect ADR?
plt.figure(figsize=(8,5))

sns.boxplot(x="market_segment",y="adr",data=df)
plt.xticks(rotation=45)
plt.title("ADR by Market Segment")
plt.show()
```

## Key Bivariate Insights

- Cancellation rate is higher for **CityHotel**, indicating higher volatility in that segment.
- Cancelled bookings generally have **higher lead times**, suggesting that early bookings are more likely to be cancelled.
- The ADR distribution shows that **[cancelled]** bookings tend to have higher prices.
- **Family bookings** show a higher cancellation rate.
- ADR varies significantly across **market segments**, with some segments consistently showing higher prices.
