```python
#Load the dataset
import pandas as pd

df = pd.read_csv("../data/excel_cleaned/hotel_bookings_cleaned.csv")
df.shape
```

```python
#Preview the dataset
df.head()
```

```python
#Check column
df.columns
```

```python
#Check data types & nulls
df.info()
```

```python
#Check missing values
df.isna().sum()
```

```python
#Fill missing values
df['children']=df['children'].fillna(0)
df['agent']=df['agent'].fillna(0)
df['country']=df['country'].fillna(0)
df['company']=df['company'].fillna(0)
```

```python
#Verify Missing Values Are Gone
df.isna().sum()
```

```python
#Fix Data Types (Important)
df['children']=df['children'].astype(int)
df['agent']=df['agent'].astype(int)
df['company']=df['company'].astype(int)

```

```python
#Fix Date Column
df['reservation_status_date']=pd.to_datetime(df['reservation_status_date'])
df.dtypes
```

```python
#Final Sanity Check
df.isna().sum()
```

```python
(df['adults']==0).sum()
((df["adr"] <= 0) & (df["is_canceled"] == 0)).sum()
((df['stays_in_week_nights']+df['stays_in_weekend_nights'])==0).sum()
df.duplicated().sum()
```

```python
#Remove exact duplicates (safe)
df=df.drop_duplicates()

#Recheck duplicates
df.duplicated().sum()

before = len(df) + 31904  # not perfect if you've already modified df; see note below
before = len(df)
df = df.drop_duplicates()
after = len(df)

before, after, before - after
```
