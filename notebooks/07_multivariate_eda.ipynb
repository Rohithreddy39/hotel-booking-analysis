# Multivariate Exploratory Data Analysis (EDA)

This notebook explores interactions between multiple variables in the hotel booking dataset.


```python
#Load the Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("../data/python_cleaned/hotel_bookings_features.csv")
df.shape
```

Multivariate Plots

```python
#1 Cancellation by Hotel Type & Lead Time

plt.figure(figsize=(8,5))

sns.scatterplot(data=df.sample(5000),x="lead_time",y="adr",hue="is_canceled",style="hotel",alpha=0.6)
plt.title("Lead Time vs ADR by cancellation and Hotel type.")

plt.show()
```

```python
#2 Cancellation Rate by Hotel & Month
monthly_cancel=df.groupby(['arrival_month','hotel'])['is_canceled'].mean().reset_index()

plt.figure(figsize=(10,5))
sns.lineplot(data=monthly_cancel,x='arrival_month',y='is_canceled',hue='hotel')
plt.title("Monthly cancellation Rate by Hotel Type.")
plt.ylabel("Cancellation Rate")

plt.show()
```

```python
#3 Revenue by Market Segment & Cancellation
plt.figure(figsize=(12,5))

sns.boxplot(data=df,x="market_segment",y="revenue",hue="is_canceled")
plt.xticks(rotation=45)
plt.title("Revenue by Market Segment and Cancellation Status")

plt.show()
```

## Key Multivariate Insights

- **Lead time is the strongest predictor of cancellations**, stronger than price (ADR) or hotel type. This indicates that bookings made far in advance carry significantly higher cancellation risk.
- **Cancellations show clear seasonality**, especially for City Hotels, suggesting that certain months or seasons are more volatile than others.
- **Revenue is concentrated in a few market segments**, particularly those driven by travel agents and tour operators.
- **High-revenue channels also carry high cancellation risk**, making these segments the most important targets for optimization strategies such as stricter cancellation policies or incentive-based retention.
