# Univariate Exploratory Data Analysis (EDA)

This notebook explores individual variables in the hotel booking dataset.

```python
#Load Feature Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("../data/python_cleaned/hotel_bookings_features.csv")

df.shape
```

First Univariate Plots

```python
#1️⃣ Cancellations distribution
sns.countplot(x="is_canceled",data=df)
plt.title("Booking Cancellation Distribution.")

plt.show()
```

```python
#2️⃣ Hotel type distribution
sns.countplot(x="hotel",data=df)
plt.title("Hotel Type Distribution.")

plt.show()
```

```python
#3️⃣ ADR (price) distribution

plt.figure(figsize=(10,5))

sns.histplot(df['adr'],bins=50,kde=True)
plt.title("ADR (Average daily rate) Distribution.")

plt.show()
```

```python
#5️⃣ Total guests distribution

sns.countplot(x="total_guests",data=df)
plt.title("Total Guests per Booking.")

plt.show()
```

```python
#1. % of bookings cancelled
cancel_counts = df['is_canceled'].value_counts(normalize=True)*100
cancel_counts
```

```python
#2. More common hotel type by bookings.
hotel_type=df['hotel'].value_counts(normalize=True)*100
hotel_type
```

```python
#3. Is ADR skewed or normal?
sns.histplot(df["adr"], bins=50, kde=True)
df['adr'].describe()
```

```python
#4. Do people book far in advance or last minute?
df["lead_time"].describe()

sns.histplot(df['lead_time'],bins=50,kde=True)
plt.title("Lead Time Distribution.")
plt.show()
(df["lead_time"] <= 30).mean() * 100
```

```python
#What is the most common group size?
df['total_guests'].value_counts().head(10)

```

## Key Univariate Insights

- Approximately **27.46%** of bookings are cancelled, indicating a **significant cancellation rate** in the dataset.
- **City Hotel** bookings are more common than **Resort Hotel** bookings, suggesting higher demand for city accommodations.
- The ADR (Average Daily Rate) distribution is **right-skewed**, meaning most bookings are concentrated at lower price ranges with a few high-priced outliers.
- Most customers book **close to the arrival date**, although a smaller portion of bookings are made far in advance.
- The most common group size is **2 guests**, indicating that the majority of bookings are likely made by couples.
